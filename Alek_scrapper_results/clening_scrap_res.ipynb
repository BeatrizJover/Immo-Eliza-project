```python
import pandas as pd
```

```python
house_df = pd.read_csv('houses_data.csv')
apartment_df = pd.read_csv('apartment_data.csv')
```

```python
house_df = house_df.assign(type_of_property=1)
house_df.tail()
```

```python
df_all_properties = pd.concat([apartment_df, house_df], ignore_index=True)

print(df_all_properties.head(10),
df_all_properties.tail(10), df_all_properties.shape)
```

```python
print(df_all_properties.duplicated(subset='immo_id').value_counts())
```

```python
df_all_properties = df_all_properties.drop(columns='immo_id')

df_all_properties.head()
```

```python
df_all_properties.to_csv('full_dataset.csv')
```
